```python
import folium

m = folium.Map(location=[-7.319456297818822, 112.73764236993138],
               zoom_start=15)

place_lat = [-7.319456297818822,-7.317987780944066,-7.317136464600442,-7.320988658098852,-7.319456297818822]
place_lng = [112.73764236993138,112.73660167284403, 112.74082883420908, 112.74185880246046,112.73764236993138]

points = []
for i in range(len(place_lat)):
    points.append([place_lat[i], place_lng[i]])

for index,lat in enumerate(place_lat):
    folium.Marker([lat, 
                   place_lng[index]],
                  popup=('patient{} \n 74contacts'.format(index)),
                 icon = folium.Icon(color='green',icon='plus')).add_to(m)
folium.PolyLine(points, color='red').add_to(m)
m
```
